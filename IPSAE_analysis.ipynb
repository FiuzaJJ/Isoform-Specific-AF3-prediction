```python
import numpy as np
import pandas as pd
import os



all_series=[]

    # Walk through all subfolders and files
for root, dirs, files in os.walk("D:\Masters_Data\Project-GPCR\IPSAE"):
        
    for file in files:
        all_series.append(pd.read_csv(f"{root}/{file}",delim_whitespace=True).iloc[2])


all_series=pd.concat(all_series,axis=1)


row=all_series.loc["Model"] # extract paths used to run IPSAE

complex_names=[]

for i in range(len(row)):
    
    complex_names.append("_".join(row.iloc[i].split("/")[-1].split("_")[2:-2]))#extract the complex names separated with _ ex.(fam64a_1_agtrap)

all_series.columns = complex_names
```

```python
substrings = ["fam64a", "nefl", "s100b","upp2"]

sub_dfs = {
    s: all_series.loc[:, all_series.columns.str.contains(s)] for s in substrings
}
```

Calculate AUCs to determine good potential candidates

```python
loaded_data = pd.read_pickle("IPSAE/IPSAE_metrics.pkl")

print(loaded_data["fam64a"])
```

```python
from sklearn.metrics import roc_curve, auc

protein_name="upp2"
df_all=pd.DataFrame(sub_dfs[protein_name])
df_all = df_all.T.dropna()

# Compute ROC curve and AUC
labels = df_all["Interaction_Found"].to_numpy()


labels = [-1 if x == "negative" else 1 for x in labels]

metrics=["ipSAE","ipSAE_d0chn","ipSAE_d0dom","ipTM_af","ipTM_d0chn","pDockQ","pDockQ2","LIS"]


metric_aucs={}
for col in metrics:
    
    fpr, tpr, thresholds = roc_curve(labels,  df_all[col].to_numpy())  

    roc_auc = auc(fpr, tpr)
    metric_aucs[col]=roc_auc


metric_aucs
```
